## C-M Interaction Curve for Rectangular Cross-Section

```python
import sympy as sy
sy.init_printing(use_latex='mathjax')
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg', 'pdf')
import numpy as np
import matplotlib.pyplot as plt
from utils import showImage
```

```python
showImage('rect-xsect')
```

Develop expressions for $C$ and $M$ as a function of $k$ ($1/2 \leq k \leq 1$).

```python
sy.var('d b C T k F_y')
```

```python
C = F_y*k*d*b
T = F_y*(1-k)*d*b
```

```python
C_r = (C-T).factor()
C_r
```

```python
M_r = (C*(d/2 - k*d/2) + T*(d/2 - (1-k)*d/2)).factor()
M_r
```

```python
C_y = F_y*b*d
M_p = F_y*b*(d/2)*(d/2)
```

```python
Cratio = C_r/C_y
Cratio
```

```python
Mratio = M_r/M_p
Mratio
```

Now plot $C/C_y$ vs $M/M_p$:

```python
# turn expressions into functions
Cfn = sy.lambdify(k,Cratio)
Mfn = sy.lambdify(k,Mratio)
```

```python
ks = np.linspace(0.5,1.0)  # vector of k values
ys = Cfn(ks)
xs = Mfn(ks)
```

```python
plt.xlabel('$M/M_p$')
plt.ylabel('$C/C_y$')
plt.grid(True)
plt.plot(xs,ys)
```
