## C-M Interaction Curves for W Shapes
Plot a $C/C_y$ vs $M/M_p$ interaction curve.

```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg', 'pdf')
import pylab
import matplotlib.pyplot as plt
```

```python
from Designer import sst
SST = sst.SST()
Fy = 345.
```

![Cross Sectional Forces](W-Shape.svg)

Calculate and return the compressive force, $C$ and the moment, $M$, corresponding to a
fully yielded cross-section with neutral axis a distance $Q$ below the centroid.
$0 \leq Q \leq D/2$

```python
def calc(B,T,W,D,Q,Fy=Fy):
    assert Q >= 0.
    assert Q <= D/2.
    H = D - 2.*T
    C = B*T*Fy          # top flange
    M = C*(H/2.+T/2.)
    if Q <= H/2.:
        f2 = Fy*W*H/2.  # web above centroid
        C += f2
        M += f2*H/4.
        f3 = Fy*W*Q     # web below centroid in compression
        C += f3
        M += -f3*Q/2.
        f4 = Fy*(H/2.-Q)*W # tension in web below centroid
        C -= f4
        M += f4*(Q+(H/2.-Q)/2.)
        f5 = Fy*B*T      # bot flange in tension
        C -= f5
        M += f5*(H/2.+T/2.)
    else:
        f2 = Fy*W*H     # total web in compression
        C += f2
        f3 = Fy*(Q-H/2.)*B  # part of flange in compression
        C += f3
        M -= f3*(H/2.+(Q-H/2.)/2.)
        f4 = Fy*(D/2.-Q)*B  # part of flange in tension
        C -= f4
        M += f4*(D/2. - (D/2.-Q)/2.)
    return C,M
```

```python
B,T,W,D,A,Zx = SST.section('W250x73',properties='B,T,W,D,A,Zx')
B,T,W,D
```

Calculate $C_y$ when N.A. is at bottom fibre, and $M_p$ when N.A. is at centroid:

```python
Cy = calc(B,T,W,D,Q=D/2.)[0]
Cy
```

```python
Mp = calc(B,T,W,D,Q=0.)[1]
Mp
```

Now check using $C_y$ and $M_p$ calculated from tabulated properties:

```python
Cyr,Mpr = Fy*A,Fy*Zx
Cyr,Mpr
```

Percent difference in each:

```python
100*(Cy-Cyr)/Cyr,100*(Mp-Mpr)/Mpr
```

As differences are 1.5%, we'll put that down to fillets, and accept it.

Now vary $Q$ from 0 to $D/2$ and generate points to plot ($M/M_p$ and $C/C_r$):

```python
cr = []
mr = []
for q in pylab.concatenate((pylab.linspace(0,D/2-T,5),pylab.linspace(D/2-T,D/2,10)[1:])):
    c,m = calc(B,T,W,D,Q=q)
    cr.append(c/Cy)
    mr.append(m/Mp)
```

Plot that interaction curve, with the accepted linear approximation:

$$
\frac{M}{M_r} = 1.18\left(1-\frac{C}{C_y}\right)
$$

```python
plt.plot(mr,cr,'+',[0.,1.18],[1,0.],'k:')
plt.xlabel('$M/M_p$')
plt.ylabel('$C/C_y$')
plt.grid(True)
plt.xlim(0,1)
```

### Compare all shapes

A function to compute the valus for all sections:

```python
def icplot(dsg):
    B,T,W,D = SST.section(dsg,properties='B,T,W,D')
    Cy = calc(B,T,W,D,Q=D/2.)[0]
    Mp = calc(B,T,W,D,Q=0.)[1]
    cr = []
    mr = []
    for q in pylab.concatenate((pylab.linspace(0,D/2-T,5),pylab.linspace(D/2-T,D/2,10)[1:])):
        c,m = calc(B,T,W,D,Q=q)
        cr.append(c/Cy)
        mr.append(m/Mp)
    plt.xlim(0,1)
    plt.ylim(0,1)
    return plt.plot(mr,cr,'+')
```

Plot a few W shapes together, of radically different sizes:

```python
icplot('W250x73')+icplot('W200x71')+icplot('W150x30')+icplot('W360x1086')+plt.plot([0,1.18],[1,0])
```

We see there is not much difference.

```python
icplot('W250x73')
```

```python
def calc_all(B,T,W,D,A,Zx,Fy=Fy):
    Cy = Fy*A
    Mp = Fy*Zx
    cr = []
    mr = []
    for q in pylab.concatenate((pylab.linspace(0,D/2-T,4),pylab.linspace(D/2-T,D/2,10)[1:])):
        c,m = calc(B,T,W,D,Q=q)
        cr.append(c/Cy)
        mr.append(m/Mp)
    return cr, mr
```

```python
N= 0
cvals = []
mvals = []
N = 0
ws = SST.section_table('W')
for i,w in ws.iterrows():
    Dsg,B,T,D,W,Use = [w[x] for x in 'Dsg,B,T,D,W,Use'.split(',')]
    if int(Use) not in [1,2]:
        continue
    if B/(2.*T) > 170/(Fy**0.5):
        continue
    H = D - 2.*T
    if H/W > 1700/(Fy**0.5):
        continue
    N += 1
    c,m = calc_all(B,T,W,D)
    cvals = cvals + c
    mvals = mvals + m
print N,'sections.'
```

```python
plt.xlim(0,1)
plt.ylim(0,1)
plt.plot(mvals,cvals,'+') + plt.plot([0,1.18],[1,0],'r',linewidth=2)
plt.xlabel('$M/M_p$')
plt.ylabel('$C/C_y$')
plt.grid(True)
```
