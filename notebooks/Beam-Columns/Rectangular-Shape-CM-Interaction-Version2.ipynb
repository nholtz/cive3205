## C-M Interaction Curve for Rectangular Cross-Section

```python
import sympy as sy
sy.init_printing(use_latex='mathjax')
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg', 'pdf')
import numpy as np
import matplotlib.pyplot as plt
from utils import showImage
```

```python
showImage('rect-xsect2')
```

Develop expressions for $C$ and $M$ as a function of $y_0 = k d$ ($0 \leq k leq 1/2$).

```python
sy.var('d b C T y_0 k, F_y')
```

```python
C_r = (y_0*b*2*F_y).subs({y_0:k*d})
C_r
```

```python
M_r = ((d/2-y_0)*b*F_y * (d/2+y_0)).subs({y_0:k*d})
M_r = M_r.expand()
M_r
```

```python
C_y = F_y*b*d
M_p = F_y*b*(d/2)*(d/2)
```

```python
Cratio = C_r/C_y
Cratio
```

```python
Mratio = M_r/M_p
Mratio = Mratio.expand()
Mratio
```

Now plot $C/C_y$ vs $M/M_p$:

```python
# turn expressions into functions
Cfn = sy.lambdify(k,Cratio)
Mfn = sy.lambdify(k,Mratio)
```

```python
ks = np.linspace(0,0.5)  # vector of k values
ys = Cfn(ks)
xs = Mfn(ks)
```

```python
plt.xlabel('$M/M_p$')
plt.ylabel('$C/C_y$')
plt.grid(True)
plt.plot(xs,ys)
```
